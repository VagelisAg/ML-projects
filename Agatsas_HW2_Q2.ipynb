```python
def plot_svm_decision_boundary(svm, X, y, title):
    plt.figure(figsize=(10, 6))
    
    # Plot data points
    for label, color, class_name in zip([0, 1], ['red', 'blue'], cancer.target_names):
        plt.scatter(X[y == label, 0], X[y == label, 1], c=color, label=class_name, edgecolors='k', alpha=0.6)
    
    # Create grid to plot decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500)) #create 2d coordinates
    grid = np.c_[xx.ravel(), yy.ravel()]
    grid_predictions = svm.decision_function(grid).reshape(xx.shape)
    
    # Plot decision boundary and margins
    plt.contour(xx, yy, grid_predictions, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='black')
    plt.title(title, fontsize=16)
    plt.xlabel("Mean Concave Points", fontsize=14)
    plt.ylabel("Worst Area", fontsize=14)
    plt.legend(title="Class", fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

# Plot decision boundary for C=0.1
plot_svm_decision_boundary(svm_clf1, X_scaled, y, "SVM Decision Boundary (C=0.1)")

# Plot decision boundary for C=1000
plot_svm_decision_boundary(svm_clf2, X_scaled, y, "SVM Decision Boundary (C=1000)")
```

```python
x0 = np.linspace(0, 5.5, 200)
pred_1 = 5 * x0 - 20
pred_2 = x0 - 1.8
pred_3 = 0.1 * x0 + 0.5

def plot_svc_decision_boundary(svm_clf, xmin, xmax):
    w = svm_clf.coef_[0]
    b = svm_clf.intercept_[0]
    x0 = np.linspace(xmin, xmax, 200)
    decision_boundary = -w[0] / w[1] * x0 - b / w[1]

    margin = 1/w[1]
    gutter_up = decision_boundary + margin
    gutter_down = decision_boundary - margin
    svs = svm_clf.support_vectors_

    plt.plot(x0, decision_boundary, "k-", linewidth=2, zorder=-2)
    plt.plot(x0, gutter_up, "k--", linewidth=2, zorder=-2)
    plt.plot(x0, gutter_down, "k--", linewidth=2, zorder=-2)
    plt.scatter(svs[:, 0], svs[:, 1], s=180, facecolors='#AAA',
                zorder=-1)
```

```python
def plot_rbf_decision_boundary(svm, X, y, title):
    plt.figure(figsize=(10, 6))
    
    # Plot data points
    for label, color, class_name in zip([0, 1], ['red', 'blue'], cancer.target_names):
        plt.scatter(
            X[y == label, 0], 
            X[y == label, 1], 
            c=color, 
            label=class_name, 
            edgecolors='k', 
            alpha=0.6
        )
    
    # Create grid for decision boundary
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500), np.linspace(y_min, y_max, 500))
    grid = np.c_[xx.ravel(), yy.ravel()]
    Z = svm.predict(grid).reshape(xx.shape)
    
    # Plot decision boundary
    plt.contourf(xx, yy, Z, alpha=0.3, levels=np.linspace(-1, 1, 3), cmap='coolwarm')
    plt.contour(xx, yy, Z, levels=[0.5], colors='black', linewidths=1.5)
    
    plt.title(title, fontsize=16)
    plt.xlabel("Mean Concave Point", fontsize=14)
    plt.ylabel("Worst Area", fontsize=14)
    plt.legend(title="Class", fontsize=12)
    plt.grid(alpha=0.5)
    plt.show()

# Plot the best RBF model decision boundary
plot_rbf_decision_boundary(best, X_scaled, y, "Best RBF SVM Decision Boundary")
```

```python
plt.figure(figsize=(10, 6))
def plot_dataset(X, y, axes):
    plt.plot(X[:, 0][y==0], X[:, 1][y==0], "bs")
    plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^")
    plt.axis(axes)
    plt.grid(True)
    plt.title("Best RBF")
    plt.xlabel("Mean Concave Point")
    plt.ylabel("Worst Area")

def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)

plot_predictions(best, [-2, 6, -2, 4])
plot_dataset(X_scaled, y, [-2, 6, -2, 4])
plt.show()
```
