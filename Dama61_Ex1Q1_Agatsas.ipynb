As we can see from the correlation heatmap of quality the features that affect quality the most is alcohol with 0.48 positive correlation and volatile acidity with 0.39 negative correlation.Also we can see that features like residual sugar free sulfur dioxide and ph have correlation close to zero which means that they do not have big impact on quality.

Q6)

```python
from sklearn.model_selection import train_test_split
```

```python
X=red_wines.iloc[:,:-1].values;X
```

```python
y=red_wines.iloc[:,-1].values;y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle=True, random_state=42,stratify=y)
```

With stratification based on quality we achieve to split the data with 80% for all classes in training set and 20% for testing.

Q7)

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train_scaled,y_train)
```

```python
predict=lm.predict(X_test_scaled);predict
```

```python
from sklearn.metrics import r2_score
print(f"{r2_score(y_test, predict):.3f}")
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, predict)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, predict)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,predict)
```

```python
accuracy=lm.score(X_test_scaled,y_test);accuracy
```

```python
plt.rc('font', size=10)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='#0343DF')
plt.scatter(y_test, predict,color="#A52A2A")
plt.title("Actuals and Predictions")
plt.legend()
plt.show()
```

As we can see the accuracy is validated from the graph as its just the 20% of the times that we actually can make a valid prediction.

Q8)

```python
from sklearn.model_selection import cross_val_score
cross_val=cross_val_score(lm, X_train_scaled, y_train, cv=10, scoring='r2');cross_val
```
